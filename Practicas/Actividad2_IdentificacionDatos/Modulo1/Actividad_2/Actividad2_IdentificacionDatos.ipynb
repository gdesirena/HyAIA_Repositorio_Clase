```python
# 0.1 Cargamos librerias de Python a utilizar
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 0.2 Cargamos libreria creada y redireccionamos path
import sys
import os

# Encontramos la librería dos niveles arriba
current_dir = os.getcwd()
root_dir = os.path.abspath(os.path.join(current_dir, '..', '..'))

# Añadir el directorio raíz al sys.path
if root_dir not in sys.path:
    sys.path.append(root_dir)

from LibreriaDatos.HyAIA import HyAIA as hia
```

```python
# 1. Carga el dataset: sns.load_dataset('tips')
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
# 2. Reporte de calidad: Usa dqr() para obtener un resumen de los datos.
cTips = hia(df_tips)
cTips.get_dqr()
```

```python
#3. Asimetría y curtósis: Calcula estos valores para cada columna numérica.
df_numericos = cTips.data_cuantitativos
df_numericos
```

```python
# Calculamos la asimetría de las columnas numéricas del DataFrame
asimetria_df = df_numericos.skew()
asimetria_df
```

```python
# Calculamos la curtosis de TODAS las columnas numéricas del DataFrame
curtosis_df = df_numericos.kurt()
curtosis_df
```

```python
df_numericos.hist()
# Podemos ver que tanto "total_bill" como "tip" tienen una asimetria positiva con cola hacia la derecha.
# Los datos de "size" son muy distantes unos de otros.
```

```python
# 4. Visualización: Grafica la distribución y medidas de forma de total_bill.
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_totalbill = df_numericos['total_bill']
df_totalbill
```

```python
# Visualización 
#--------------------

# Configurar el estilo de los gráficos
plt.figure(figsize=(6, 4)) 

# Subplot 1: Histograma de la distribución
plt.subplot(2, 1, 1) 
sns.histplot(df_totalbill, kde=True, color='skyblue', bins=20) # bins para el número de barras
plt.title(f'Distribución de (total_bill)')
plt.xlabel('') 
plt.ylabel('Frecuencia')

# Subplot 2: Boxplot para mostrar cuartiles y outliers
plt.subplot(2, 1, 2) 
sns.boxplot(x = df_totalbill, color='coral', width=0.5)
plt.title(f'Medidas de forma de (total_bill)')
plt.xlabel('total_bill')
plt.ylabel('') 

plt.tight_layout() # Ajustar parámetros
plt.show()
```

```python
# 5. Comparación por día: Compara asimetría y curtósis de total_bill por day.

from scipy.stats import kurtosis

df_base = cTips.data[['total_bill', 'day']].copy()
medidas_forma_por_dia = df_base.groupby('day').agg(
    asimetria=('total_bill', 'skew'),
    curtosis=('total_bill', kurtosis) 
)
medidas_forma_por_dia
```

```python
# 6. Cuartiles: Calcula los cuartiles de tip.
tips = df_numericos['tip']
tips
```

```python
q1 = np.percentile(tips, 25)
q2 = np.percentile(tips, 50)  # Mediana
q3 = np.percentile(tips, 75)
print('\nCualtiles de la columna tip:')
print('Q1 (25%):', q1)
print('Q2 (Mediana, 50%):', q2)
print('Q3 (75%):', q3)
```

```python
# 7. Diagrama de cajas: Visualiza tip y comenta tus observaciones.

# Configurar el estilo de los gráficos
plt.figure(figsize=(6, 4)) 

plt.subplot(2, 1, 1) 
sns.histplot(tips, kde=True, color='skyblue', bins=20) # bins para el número de barras
plt.title(f'Distribución de (tip)')
plt.xlabel('') 
plt.ylabel('Frecuencia')

# Boxplot para mostrar cuartiles y outliers
plt.subplot(2, 1, 2) 
sns.boxplot(x = tips, color='coral', width=0.5)
plt.title(f'Medidas de forma para (Tip)')
plt.xlabel('Tips')
plt.ylabel('') 

plt.tight_layout() # Ajustar parámetros
plt.show()
```

```python
# 1.  utiliza el dataset credit_card.csv.

df_creditcard = pd.read_csv('credit_card.csv')
df_creditcard.head()
```

```python
# Identificación de outliers multivariados: Aplica técnicas multivariadas vistas en clase para detectar valores atípicos.

# Obtenemos información de las columnas de salaries
cCreditcard = hia(df_creditcard)
cCreditcard.get_dqr()
```

```python
# Obtenemos las columnas cuantitativas.
df_cuantitativos = cCreditcard.data_cuantitativos
df_cuantitativos
```

```python
# Completamos nulos con la mediana para tener datos en todos los registros.
df_cuantitativos = df_cuantitativos.fillna(df_cuantitativos.median())
df_cuantitativos
```

```python
from sklearn.cluster import DBSCAN
X = df_cuantitativos[['Family_Members','Annual_income']].values

db = DBSCAN(eps=3.0, min_samples=10).fit(X)
labels = db.labels_
labels
```

```python
pd.Series(labels).value_counts() #-1 representan valores atipicos
```

```python
unique_labels = set(labels)
colors = ['blue', 'red']

plt.figure(figsize=(6,4))
for color,label in zip(colors, unique_labels):
    sample_mask = [True if l == label else False for l in labels]
    plt.plot(X[:,0][sample_mask], X[:, 1][sample_mask], 'o', color=color);
plt.xlabel('Family_Members');
plt.ylabel('Annual_income');
plt.show()
```
