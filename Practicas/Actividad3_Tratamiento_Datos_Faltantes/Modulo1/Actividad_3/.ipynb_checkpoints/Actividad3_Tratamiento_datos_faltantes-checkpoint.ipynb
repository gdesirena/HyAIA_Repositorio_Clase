### Llenado con valores vecinos

**1.6.-El comando `fillna()` tiene el argumento `method` que permite hacer un llenado hacia atrás ('bfill') o hacia adelante ('ffill'). Utilice `ffill` y `bfill` con el método `fillna` para el dataframe `df_copy` de los índices 25 hasta 30. Tenga en cuenta que al usar ésta opción si el dato con el que se intenta hacer el llenado también es un NaN, éste permanece.**

```python
# Aplicamos Forward Fill (ffill) al rango de registros especificado
# Los valores nulos se llenan con el valor anterior no nulo.
df_copy.loc[25:30] = df_copy.loc[25:30].fillna(method='ffill')

# Aplicamos Backward Fill (bfill) al rango de registros especificado
# Los valores nulos se llenan con el valor posterior no nulo.
df_copy.loc[25:30] = df_copy.loc[25:30].fillna(method='bfill')
```

```python
# Visualizamos el grafico para ver como cambia
msno.matrix(df_copy)
```

### Llenado de datos usando sklearn



**1.7.- Es posible también usar el metodo 'SimpleImputer' de la libreria sklearn para hacer llenado, éste nos permite definir cual estrategia usar para el llenado de los datos (media, mediana o moda). importe la clase `SimpleImputer` de `sklearn.impute`. Aplique el llenado de los datos nulos mediante la estrategia 'most_frequent'**

```python
# Importamos la clase SimpleImputer
from sklearn.impute import SimpleImputer
```

```python
# Imputamos los datos faltantes con la moda (most_frecuent)
imp_mode = SimpleImputer(strategy='most_frequent')
df_imputed = pd.DataFrame(imp_mode.fit_transform(df_copy), columns=df_copy.columns)
df_imputed
```

```python
# Nuevamente visualizamos el grafico para ver como cambia.   Ahora en este DataSet no hay nulos.
msno.matrix(df_imputed)
```

#### Tratamiento de los datos faltantes

Solo debemos eliminar una columna si la mayoria de entradas son vacias. En nuestro caso, ninguna de las columnas cumplen con este criterio como para ser eliminadas. 

En este caso tenemos cierta libertad para aplicar diferences métodos de llenado de datos faltantes; no obstante, algunos métodos pueden resultar más adecuados que otros. Aplique los siguientes métodos para cada columna:



**2.4.- Reemplazo por la media:**
<ul>
    <li>"normalized-losses"</li>
    <li>"stroke"</li>
    <li>"bore"</li>
    <li>"horsepower"</li>
    <li>"peak-rpm"</li>
</ul>

**2.5.- Replace por la moda:**
<ul>
    <li>"num-of-doors" 
        <ul>
            <li>Razón: 84% de los sedan son 4 puertas. Dado que los carros cuatro puertas son los más frecuentes, la probabilidad de ocurrencia es mayor.</li>
        </ul>
    </li>
</ul>

**2.5.- Eliminación de una fila entera:**
<ul>
    <li>"price": 4 datos faltantes
        <ul>
            <li>Razón: Como, eventualmente, el precio es algo que se puede predecir, ninguna entrada sin precio puede ser usada para una predicción; por ende, cualquier fila sin precio no será útil.</li>
        </ul>
    </li>
</ul>

```python
# 2.4 Remplazaremos por la media las columnas especificadas.

# Definimos las columnas a imputar con la media
columnas_a_imputar = ['normalized-losses', 'stroke', 'bore', 'horsepower', 'peak-rpm']
#columnas_a_imputar = 'normalized-losses'

#media_col = cAutos.data[columnas_a_imputar].mean()
#print(f"La media de la columna {columnas_a_imputar} es {media_col} ")


for col in columnas_a_imputar:
    # Convertimos a tipo de dato numérico
    cAutos.data[col] = pd.to_numeric(cAutos.data[col], errors='coerce')

    # Obtenemos la media para cada columna
    #media_col = cAutos.data[col].mean(numeric_only=True)
    #print(f"La moda de la columna {col} es {moda_col} ")
    
    cAutos.data[col] = cAutos.data[col].fillna(
        cAutos.data[col].mean()
    )   
```

```python
# 2.5 Remplazaremos por la moda la columna "num-of-doors".

# Definimos las columnas a imputar con la moda
columnas_a_imputar = ['num-of-doors']

for col in columnas_a_imputar:
    
    # Convertimos a tipo de dato a String
    cAutos.data[col] = cAutos.data[col].astype(str)
    
    # Aplicamos la moda a los datos faltantes
    cAutos.data[col] = cAutos.data[col].fillna(
        cAutos.data[col].mode()
    ) 
```

```python
# 2.5 Eliminación de una fila entera.
cAutos.data.shape
```

```python
# Eliminamos los registros cuando el valor de la columna price es vacio
cAutos.data.dropna(subset=['price'], inplace=True)
```
