**1.- Modificar la funci√≥n dqr() para que en el reporte se agregue una columna que sea booleana boolena para representar si una columna es categ√≥rica ('Is_categorical') ('Is_categorical' True-->col categorica, False ---> num√©rica)**


```python
# Importar las librer√≠as necesarias
import pandas as pd
import numpy as np
from HyAIA import HyAIA

# Cargar el dataset real de viviendas en USA
print("Cargando dataset USA_Housing.csv...")
df_housing = pd.read_csv('data/USA_Housing.csv')

print(f"\n‚úì Dataset cargado: {df_housing.shape[0]} filas, {df_housing.shape[1]} columnas")
print("\nPrimeras 5 filas del dataset:")
df_housing.head()
```

```python
# PUNTO 1: Verificar la columna 'Is_categorical'
# Crear instancia de HyAIA con el dataset real
print("Generando reporte de calidad de datos con HyAIA...\n")
hyaia = HyAIA(df_housing)

# Obtener el reporte de calidad de datos
reporte = hyaia.df_dqr

print("=" * 90)
print("PUNTO 1: Columna 'Is_categorical' agregada")
print("=" * 90)
print("\nVerificaci√≥n de tipos de columnas:")
print(reporte[['Columns_Names', 'Dtypes', 'Is_categorical']])
print("\n‚úÖ Las columnas num√©ricas tienen Is_categorical = False")
print("‚úÖ Las columnas de texto (Address) tienen Is_categorical = True")
```
