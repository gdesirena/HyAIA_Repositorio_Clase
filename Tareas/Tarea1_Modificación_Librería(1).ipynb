**1.- Modificar la función dqr() para que en el reporte se agregue una columna que sea booleana boolena para representar si una columna es categórica ('Is_categorical') ('Is_categorical' True-->col categorica, False ---> numérica)**


```python
import pandas as pd

def dqr(df):
    """
    Genera un reporte de calidad de datos con información básica de cada columna.
    Incluye una columna booleana 'Is_categorical' que indica si la columna es categórica.
    """
    # Detectamos columnas categóricas (object o category)
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns
    
    # Construimos el reporte
    reporte = pd.DataFrame({
        "Type": df.dtypes,
        "Missing_Values": df.isnull().sum(),
        "Missing_%": df.isnull().mean() * 100,
        "Unique_Values": df.nunique(),
        "Is_categorical": [col in categorical_cols for col in df.columns]
    })
    
    return reporte
```

**2.- Modificar la función dqr() para que en el reporte se agregue una columna con las categorías de las columnas (valores unicos) que resulten ser categóricas (si las categorias > 10, no agregar esas categorias)**


```python
import pandas as pd

def dqr(df):
    """
    Genera un reporte de calidad de datos con información básica de cada columna.
    Incluye:
    - Is_categorical: True si la columna es categórica.
    - Categories: lista de categorías si la columna es categórica y tiene <= 10 valores únicos.
    """
    # Detectamos columnas categóricas (object o category)
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns

    # Construimos la lista de categorías
    categories_list = []
    for col in df.columns:
        if col in categorical_cols:
            unique_vals = df[col].dropna().unique()
            if len(unique_vals) <= 10:
                categories_list.append(list(unique_vals))
            else:
                categories_list.append("Más de 10 categorías")
        else:
            categories_list.append(None)

    # Construimos el reporte final
    reporte = pd.DataFrame({
        "Type": df.dtypes,
        "Missing_Values": df.isnull().sum(),
        "Missing_%": df.isnull().mean() * 100,
        "Unique_Values": df.nunique(),
        "Is_categorical": [col in categorical_cols for col in df.columns],
        "Categories": categories_list
    })

    return reporte

```

**3.- Aplicar el análisis estadístico de las columnas creadas (max, min, mean, std) solo para las columnas que sean numéricas ('Is_categorical' == False)**


```python
import pandas as pd

def dqr(df):
    """
    Genera un reporte de calidad de datos con información básica de cada columna.
    Incluye:
    - Identificación de columnas categóricas (Is_categorical).
    - Categorías (solo si <= 10 únicas).
    - Estadísticos básicos para columnas numéricas (Min, Max, Mean, Std).
    """
    # Detectar columnas categóricas
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns
    
    # Lista de categorías (solo si <= 10 únicas)
    categories_list = []
    for col in df.columns:
        if col in categorical_cols:
            unique_vals = df[col].dropna().unique()
            if len(unique_vals) <= 10:
                categories_list.append(list(unique_vals))
            else:
                categories_list.append("Más de 10 categorías")
        else:
            categories_list.append(None)

    # Estadísticos básicos para columnas numéricas
    min_list, max_list, mean_list, std_list = [], [], [], []
    for col in df.columns:
        if col not in categorical_cols:  # Es numérica
            min_list.append(df[col].min())
            max_list.append(df[col].max())
            mean_list.append(df[col].mean())
            std_list.append(df[col].std())
        else:
            min_list.append(None)
            max_list.append(None)
            mean_list.append(None)
            std_list.append(None)

    # Construcción del reporte final
    reporte = pd.DataFrame({
        "Type": df.dtypes,
        "Missing_Values": df.isnull().sum(),
        "Missing_%": df.isnull().mean() * 100,
        "Unique_Values": df.nunique(),
        "Is_categorical": [col in categorical_cols for col in df.columns],
        "Categories": categories_list,
        "Min": min_list,
        "Max": max_list,
        "Mean": mean_list,
        "Std": std_list
    })

    return reporte

```
