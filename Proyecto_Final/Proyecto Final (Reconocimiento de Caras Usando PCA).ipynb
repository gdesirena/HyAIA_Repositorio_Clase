<span style="font-size:1.1em;"><b>Descripción:</b> Desarrolla un modelo de IA capaz de reconocer caras usando PCA. </span>
<ul style="margin-top:10px;">
</ul>
</div>

<div style="margin-top:18px; background:#f0f7ff; border-radius:8px; padding:14px; border:1px solid #b3c6ff;">
<b>Requerimientos:</b>
<ul style="margin-bottom:0;">
<li>Recopilación de datos: Reúne un conjunto diverso de imágenes faciales, asegurándote de que sean de alta calidad y variadas en iluminación, expresiones y fondos. Dataset recomendado: <a href="https://www.cl.cam.ac.uk/research/dtg/attarchive/facedatabase.html" target="_blank">FacesDatabase</a> </li>
<li>Preprocesamiento: Preprocesa las imágenes usando libreias como matplotlib, open cv, asegurandose que se tengan imagenes del mismo tamaño.</li>
<li>Reducción de dimensionalidad: Aplica PCA a las imágenes preprocesadas para reducir su dimensionalidad, preservando las características más significativas.</li>
<li>Extracción de características: Utiliza las Eigenfaces (eigenvectores de las imagenes preprocesadas) extraídas para crear un espacio de características que pueda emplearse en tareas de reconocimiento facial.</li>
<li>Pruebas: Valida el modelo utilizando un conjunto de datos de prueba independiente para evaluar su precisión y rendimiento.
</ul>
</div>
