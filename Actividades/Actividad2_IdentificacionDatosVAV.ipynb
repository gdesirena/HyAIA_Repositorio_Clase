```python
#  Importamos las librerías
# - pandas: para manipular datos en tablas (DataFrames)
# - matplotlib y seaborn: para gráficas


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict


# ---------- Definir dqr() ----------
def dqr(df: pd.DataFrame) -> pd.DataFrame:
    """
    Data Quality Report (DQR)
    Devuelve: Type, N_Null, Pct_Null, N_Unique, Is_categorical
    """
    # Esta función arma un “reporte de salud” del DataFrame:
    # te dice tipo de dato por columna, cuántos nulos, porcentaje de nulos,
    # cuántos valores únicos y si la columna es categórica.
    report = pd.DataFrame({
        "Type": df.dtypes.astype(str),              # Tipo de dato (int, float, object, category, etc.)
        "N_Null": df.isnull().sum(),                # Cuántos nulos hay por columna
        "Pct_Null": (df.isnull().mean() * 100).round(2),  # % de nulos por columna, con 2 decimales
        "N_Unique": df.nunique(dropna=True)         # Conteo de valores únicos (sin contar NaN)
    })
    # Aquí marcamos si la columna es “categórica” a ojo de buen cubero:
    # consideramos categórica si es tipo object o category.
    report["Is_categorical"] = report["Type"].apply(
        lambda t: True if ("object" in t or "category" in t) else False
    )
    report.index.name = "column"   # Renombramos el índice para que salga bonito en el reporte
    return report                  # Regresamos el DataFrame con el diagnóstico

# ---------- Cargar dataset (robusto) ----------
try:
    # Primer intento: cargar el dataset "tips" directo de seaborn (si hay internet y la librería tiene el dataset)
    tips = sns.load_dataset("tips")
    print("Dataset 'tips' cargado desde seaborn.")
except Exception as e:
    # Si falla (por ejemplo, sin red), intentamos leer un CSV local llamado tips.csv
    print("No fue posible cargar desde seaborn (posible falta de red). Intentando 'tips.csv'...")
    try:
        tips = pd.read_csv("tips.csv", parse_dates=False)
        print("Dataset 'tips' cargado desde tips.csv en carpeta local.")
    except Exception as e2:
        # Si tampoco está el CSV, no nos quedamos a medias: armamos un mini dataset casero para practicar.
        print("No hay 'tips.csv' en la carpeta. Se usará un ejemplo reducido para practicar.")
        tips = pd.DataFrame({
            "total_bill": [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77],
            "tip": [1.01, 1.66, 3.5, 3.31, 3.61, 4.71, 2.0],
            "sex": ["Female","Male","Male","Male","Female","Male","Male"],
            "smoker": ["No","No","No","No","No","No","No"],
            "day": ["Sun","Sun","Sun","Sun","Sun","Sun","Sun"],
            "time": ["Dinner","Dinner","Dinner","Dinner","Dinner","Dinner","Dinner"],
            "size": [2,3,3,2,4,4,2]
        })
        # Este dataset de prueba trae lo básico: cuenta total, propina, sexo, si fuma,
        # día, tiempo (comida/cena), y tamaño de mesa. Suficiente para correr el DQR y jugar con gráficos.

# Ejemplo de uso (si quieres ver el reporte de calidad de datos):
# print(dqr(tips))
# También podrías graficar algo rápido:
# sns.scatterplot(data=tips, x="total_bill", y="tip")
# plt.show()
```

```python

#  1 Reporte de calidad
report = dqr(tips)
print("\n--- DQR ---")
print(report)
```

```python

# 2. Skewness y Kurtosis para columnas numéricas
numeric_cols = tips.select_dtypes(include=[np.number]).columns.tolist()
skewness = tips[numeric_cols].skew().rename("skewness")
kurtosis = tips[numeric_cols].kurtosis().rename("kurtosis")
shape_stats = pd.concat([skewness, kurtosis], axis=1)
print("\n--- Skewness y Kurtosis (columnas numéricas) ---")
print(shape_stats)
```

```python
#  3. Visualización: distribución de total_bill
import numpy as np
import matplotlib.pyplot as plt

# 3. Visualización: distribución de total_bill en color rosa
if "total_bill" in tips.columns:
    total = tips["total_bill"].dropna()

    plt.figure(figsize=(8, 5))

    # --- Histograma en rosa ---
    counts, bins, patches = plt.hist(
        total,
        bins=25,
        color="#EC407A",      # rosa para las barras
        edgecolor="white",    # borde blanco para que se vean limpias
        alpha=0.9
    )

    # --- Líneas de referencia (media/mediana/moda-bin) en rosa oscuro ---
    mean_val   = total.mean()
    median_val = total.median()

    # Moda aproximada: centro del bin con mayor frecuencia
    mode_bin_idx = np.argmax(counts)
    # centro del bin dominante
    mode_bin_center = (bins[mode_bin_idx] + bins[mode_bin_idx + 1]) / 2.0

    # Líneas verticales
    plt.axvline(mean_val,   color="#D81B60", linestyle="--", linewidth=2,
                label=f"mean={mean_val:.2f}")
    plt.axvline(median_val, color="#D81B60", linestyle="-.", linewidth=2,
                label=f"median={median_val:.2f}")
    plt.axvline(mode_bin_center, color="#D81B60", linestyle=":", linewidth=2,
                label=f"mode_bin={mode_bin_center:.2f}")

    # Títulos y ejes (también en rosa oscuro para coherencia visual)
    plt.xlabel("total_bill", color="#D81B60")
    plt.ylabel("frequency", color="#D81B60")
    plt.title("Histograma de total_bill", color="#D81B60")
    plt.legend()

    # Color en ticks y spines para el look & feel
    ax = plt.gca()
    ax.tick_params(colors="#D81B60")
    ax.spines["bottom"].set_color("#D81B60")
    ax.spines["left"].set_color("#D81B60")

    plt.tight_layout()
    plt.show()
else:
    print("La columna 'total_bill' no está presente en el DataFrame.")
```

```python
# 4. Comparación por día
group_shape = tips.groupby("day")["total_bill"].agg(
    count="count",
    mean="mean",
    median="median",
    skew=lambda x: x.skew(),
    kurtosis=lambda x: x.kurtosis()
)

print("\n--- total_bill: shape por day ---")
print(group_shape)

```

```python

# 5. Cuartiles de tip


if "tip" in tips.columns:
    tip_quartiles = tips["tip"].quantile([0, 0.25, 0.5, 0.75, 1]).rename(index={0:"min", 0.25:"Q1", 0.5:"median", 0.75:"Q3", 1:"max"})
    print("\n--- Cuartiles de 'tip' ---")
    print(tip_quartiles)
else:
    print("\nLa columna 'tip' no está disponible en el dataset de ejemplo.")
```

```python

#  6. Boxplot de tip
if "tip" in tips.columns:
    plt.figure(figsize=(6,4))
    plt.boxplot(tips["tip"].dropna(), vert=True, tick_labels=["tip"])
    plt.title("Boxplot de 'tip'")
    plt.ylabel("Cantidad de propina")
    plt.show()
else:
    print("No hay columna 'tip' para boxplot.")


print("1) DQR: revisa tipos y nulos. Marca las columnas categóricas (Is_categorical True/False).")
print("2) Forma:")

# Define the variables before use
if "total_bill" in tips.columns:
    mean_total = tips["total_bill"].mean()
    median_total = tips["total_bill"].median()
    skew_total = tips["total_bill"].skew()
    kurt_total = tips["total_bill"].kurtosis()
else:
    mean_total = None
    median_total = None
    skew_total = None
    kurt_total = None

if mean_total is not None:
    print(f"   - total_bill: mean={mean_total:.2f}, median={median_total:.2f}, skew={skew_total:.2f}, kurtosis={kurt_total:.2f}.")
    print("   - Interpretación típica: skew>0 sugiere sesgo a la derecha (colas largas), kurtosis>0 sugiere colas pesadas.")
else:
    print("   - No se dispone de total_bill en el dataset de ejemplo para dar medidas.")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Estilo general y paleta en morados
sns.set_theme(style="whitegrid")
PURPLE      = "#7E57C2"  # principal
DEEP_PURPLE = "#5E35B1"  # contraste

# Si df ya está definido, seleccionamos columnas numéricas
num_cols = df.select_dtypes(include=['number']).columns

for col in num_cols:
    plt.figure(figsize=(12, 4))

    # -------- Histograma + KDE (todo en morado) --------
    plt.subplot(1, 2, 1)
    sns.histplot(
        df[col].dropna(),
        kde=True,
        color=PURPLE,                 # barras moradas
        edgecolor="white",
        alpha=0.85,
        line_kws={"color": DEEP_PURPLE, "linewidth": 2}  # línea KDE morado profundo
    )
    plt.title(f"Distribución de {col}", color=DEEP_PURPLE)
    plt.xlabel(col, color=DEEP_PURPLE)
    plt.ylabel("Frecuencia", color=DEEP_PURPLE)
    ax1 = plt.gca()
    ax1.tick_params(colors=DEEP_PURPLE)
    ax1.spines["bottom"].set_color(DEEP_PURPLE)
    ax1.spines["left"].set_color(DEEP_PURPLE)

    # -------- Boxplot (caja y detalles en morado) --------
    plt.subplot(1, 2, 2)
    boxprops     = {"facecolor": PURPLE, "edgecolor": DEEP_PURPLE, "linewidth": 1.5}
    medianprops  = {"color": DEEP_PURPLE, "linewidth": 2}
    whiskerprops = {"color": DEEP_PURPLE, "linewidth": 1.5}
    capprops     = {"color": DEEP_PURPLE, "linewidth": 1.5}
    flierprops   = {"marker": "o", "markerfacecolor": DEEP_PURPLE,
                    "markeredgecolor": DEEP_PURPLE, "alpha": 0.7}

    sns.boxplot(
        x=df[col].dropna(),
        color=PURPLE,                 # color base de la caja
        boxprops=boxprops,
        medianprops=medianprops,
        whiskerprops=whiskerprops,
        capprops=capprops,
        flierprops=flierprops
    )
    plt.title(f"Diagrama de Caja de {col}", color=DEEP_PURPLE)
    plt.xlabel(col, color=DEEP_PURPLE)
    ax2 = plt.gca()
    ax2.tick_params(colors=DEEP_PURPLE)
    ax2.spines["bottom"].set_color(DEEP_PURPLE)
    ax2.spines["left"].set_color(DEEP_PURPLE)

    plt.tight_layout()
    plt.show()
```

```python
#2 Identificación de outliers
num_cols = df.select_dtypes(include=['int64','float64']).columns


def iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return series[(series < lower) | (series > upper)]

for col in num_cols:
    outliers = iqr_outliers(df[col])
    print(f"{col}: {len(outliers)} outliers")
```

```python

from scipy.stats import zscore

for col in num_cols:
    zs = zscore(df[col])
    outliers = df[abs(zs) > 3][col]
    print(f"{col}: {len(outliers)} outliers (Z-score)")
```
